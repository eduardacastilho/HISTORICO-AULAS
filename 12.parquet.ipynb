```python
hora_inicio = dt.datetime.now()

lista_arquivos = os.listdir(DADOS)
print(lista_arquivos)
for arquivo in lista_arquivos: 
    print(f'Lendo arquivo: {arquivo}')
    df = pl.read_csv(DADOS+arquivo, separator=';', encoding='iso-8859-1')
    #se existe df_bf nas variaveis locais
    if 'df_bf' in locals():
        df_bf= pl.concat([df_bf, df])
    else:
        df_bf=df
    del df 

  

```

```python
df_bf = df_bf.with_columns(
        pl.col('VALOR PARCELA').str.replace(',', '.').cast(pl.Float64)
)

df_bf.write_parquet('C:\analise de dados\arquivosBF\dados_bronze\df_bf.parquet')
```
